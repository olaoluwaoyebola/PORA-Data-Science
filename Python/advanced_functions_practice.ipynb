```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def next_year_age(self):
        next_year_age = self.age + 1
        print(f"You'll be {next_year_age} years old by next year")
```

```python
p1 = Person("Isaac", 25)
p2 = Person("John", 28)
```

```python
p1.next_year_age()
```
