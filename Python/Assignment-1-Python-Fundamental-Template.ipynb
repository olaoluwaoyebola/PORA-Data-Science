## 4. While Loop with Break and Pass

```python
# Task: Use a while loop to print numbers from 1 to 10.
# If the number is 5, use pass (do nothing).
# If the number is 8, stop the loop using break.
```

```python
""" Write your code hereðŸ‘‡ """
numbers = 1    # current number

while numbers <= 10:    # stating the limit of numbers needed
    print(numbers)
    numbers += 1        # adds 1 to the current number till the limit is reached
    if numbers == 5:
        pass            # do nothing if the number equals 5
    elif numbers == 8:
        print(f"Eighth digit reached, breaking loop")
        break           # break the loop once the number equals 8
print("Loop broken")
```

## 5. Combined Logic (List, Condition, and String Formatting)

```python
# Task: Create a list of student scores: [45, 67, 89, 32, 76].
# Using a for loop and if condition, 
# print "Pass" for scores â‰¥ 50 and "Fail" for scores < 50, formatted as:
# "Score: 67 - Pass"
```

```python
"""Write your code hereðŸ‘‡"""

student_scores = [45, 67, 89, 32, 76]    # list of student scores
for score in student_scores:
    if score >= 50:                      # checks for scores from 50 and above
        print(f"Score: {score} - Pass")
    else:                                # checks for scores below 50
        print(f"Score: {score} - Fail")
```

## 8. Map, Filter, and Reduce

```python
# Task:
# Given a list of numbers [3, 7, 10, 14, 18]:
# - Use map() to convert each number to its square
# - Use filter() to select only numbers greater than 100
# - Use reduce() to compute the sum of the filtered list
```

```python
"""Write your code hereðŸ‘‡"""

from functools import reduce

num_list2 = [3, 7, 10, 14, 18]

squared_num = map(lambda num: num ** 2, num_list2)
num_above_100 = filter(lambda num: num > 100, squared_num)
num_total = reduce(lambda x, y: x + y, num_above_100)

num_total
```

## 9. Function and User Input

```python
# Task: Import the math module and create a function named circle_area() 
#          that takes the radius as an argument and returns the area of the circle using the formula Ï€rÂ².
# Use the input() function to collect the radius value from the user, call the function, 
#           and print the result formatted as: "The area of the circle is 78.54"
```

```python
"""Write your code hereðŸ‘‡"""

import math as m

def circle_area(radius):
    #if radius <0:
        #print("Radius must be a positive value")
    # area = Ï€rÂ²
    area = m.pi * (radius ** 2)
    print(f"The area of the circle is {area}")

radius = float(input("Enter the radius value: "))
circle_area(radius)
```
