```python
import pandas as pd
```

```python
data = {
    'Fruits': ['Apple', 'Grape', 'Pear', 'Banana', 'Orange'],
    'Price': [400, 500, 2000, 300, 250],
    'Quantity': [5,10, 6, 8, 12]
}

df = pd.DataFrame(data)
df = df.set_index('Fruits')
df
```

```python
#loc
df.loc[['Apple', 'Grape']]
```

```python
# iloc: uses index position
df.iloc[[0,1]]
```

```python
# select some or all row and a column

df.loc[ : ,['Price']]
```

```python
## print out the index, column of df

df.columns
df.index
```

```python
# using np to get sqrt

import numpy as np
np.sqrt(df['Price'])
```

```python
df['Price'].mean()
```

```python
# getting mean of all columns
df.mean(axis = 0)
```
