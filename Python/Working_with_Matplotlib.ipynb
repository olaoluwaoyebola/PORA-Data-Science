## Working with Matplotlib

#### Matplotlib is a powerful Python library for creating static, animated, and interactive charts

### Why Matplotlib

- Customization: Control over every aspect of a plot, including colors, labels, and styles.

- Integration: Works well with NumPy, Pandas, and Jupyter Notebooks.

- Multiple Interfaces: The Pyplot module (matplotlib.pyplot) provides a MATLAB-like interface, while the object-oriented API allows for more complex customizations.

### Working with Figures and Axes in Matplotlib

#### Matplotlib provides two main interfaces for plotting:
  
- Pyplot API (State-Based Approach): Uses plt.plot() directly (simpler, but less flexible).

- Object-Oriented API (recommended for complex plots): Uses Figure and Axes objects for more control.

  
### Anatomy and Customization of a Matplotlib Plot

- Figure: The overall container or canvas for the plot.

- Axes: The X and Y axes that define the plotâ€™s coordinate system.

- Title: A descriptive label for the plot.

- Legend: Explains symbols or colors used in the plot.

- Ticks: Small markers along the axes indicating specific values.

- Grids: Background lines that help align values on the axes.

- Lines/Markers: Represent the actual data, with lines for continuous values and markers for discrete points.

### Basic Plot types in Matplotlib

- Line Plot: plt.plot()

- Scatter Plot: plt.scatter()

- Bar Chart: plt.bar()

- Histogram: plt.hist()

- Pie Chart: plt.pie()

### Object Oriented Method

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure(figsize=(6,4))
ax = fig.add_axes([1,1,1,1])

#left, bottom, width, height

ax.set_title('Figure with One Axes')
ax.set_xlabel('X axis')
ax.set_ylabel('Ylabel')

#ax.legend()
plt.show();
```

```python
fig = plt.figure(figsize=(8,6))

ax1 = fig.add_axes([0.1,0.1,0.8,0.8])
ax1.plot([1,2,3,4], [10,20,15,25], 'g', label ='Main Plot')
ax1.set_title('Main axes')
plt.grid()

#left, bottom, width, height

ax1 = fig.add_axes([0.2,0.2,0.3,0.3])
ax1.plot([1,2,3,4], [5,15,10,20],'r', label ='Inset Plot')
ax1.set_title('Inset axes')

plt.grid()
plt.show()
```

### Subplots

```python
fig, axes = plt.subplots(2, 2, figsize=(8,6))
axes[0,0].plot([1,2,3,4], [10,20,15,25], label="Example")

axes[0,0].set_title('Using plt.subplots()')
axes[0,0].set_xlabel('X axis')
axes[0,0].set_ylabel('Y label')
axes[0,0].legend()
plt.show()
```

### Creating a subplots

```python
fig, axes = plt.subplots(2,2, figsize=(8,6))

axes[0,0].plot([1,2,3], [10,20,30])
axes[0,0].set_title('Plot 1')

axes[0,1].bar(["A", "B", "C"], [5,15,10])
axes[0,1].set_title('Plot 2')

axes[1,0].scatter([1,2,3], [30,10,20])
axes[1,0].set_title('Plot 3')

axes[1,1].hist([2,2,2,3,3,4,4,4,5,5,5,5])
axes[1,1].set_title('Plot 4')





plt.tight_layout()
plt.show()
```

```python
data = {'Cashew nut' : 800,
     'Tiger nut' : 500,
     'Ground nut' :400,
     'Bambara nut': 300,
     'Wal nut': 250}

#Setting up the plot
fig, ax = plt.subplots()

#PLotting the chart
ax.barh(y = list(data.keys()), width = list(data.values()))

#Customize the chart
ax.set(title = 'Chart of nuts',
       xlabel = 'Prices',
       ylabel = 'Nuts')


plt.show()

fig.savefig('Nut Chart')
```

### Plotting with Pandas

```python
#Importing of library
import pandas as pd

#Preparing of data
df = pd.read_csv('13.1 heart-disease.csv')
df.head()
```

```python
#Setting up and plotting of data

df['age'].plot.hist(bins=10)

plt.show()
```

### Example 4

#### Object Oriented method with Pyplot

```python
#Preparing of data

over_50  = df[df['age']>50]
over_50.head()
```

```python
#Setting up plot

fig, ax = plt.subplots()
over_50.plot(kind='scatter',
             x ='age',
             y ='chol',
             c ='target',
             cmap = 'winter',
             ax=ax)
plt.show()
```

### Example 5

#### Object Oriented from Scratch

```python
#Seting up of the data
fig, ax = plt.subplots()

#plotting of data
scatter  = ax.scatter(x=over_50['age'],
                      y = over_50['chol'],
                      c=over_50['target'],
                      cmap = 'plasma')

#customizing plot
ax.set(title = "Heart Disease and cholestorel levels",
       xlabel  = 'Age',
       ylabel = 'cholesterol')
# Adding a legend
ax.legend(*scatter.legend_elements(), title ='targets')

plt.show()
```
