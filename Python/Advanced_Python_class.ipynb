## Map Function 

The map function applies a given function to all items in the collection.

Formula for map function = map(function, collection)

```python
#Example 1 on map function
celsius_temp = [0.0, 10.0, 20.0, 25.0, 30.0, 40.0, 50.0]

def c_to_f(temp):
    return (temp *9/5)+32

#function = c_to_f ✅
#collection = celsius_temp ✅

fahrenheit_temp  = list(map(c_to_f, celsius_temp))
print(f'Temperature in celsius is: {celsius_temp}')
print(f'Temperature in fahrenheit is: {fahrenheit_temp}')
```

```python
celsius_temp = [0.0, 10.0, 20.0, 25.0, 30.0, 40.0, 50.0]

def c_to_f(temp):
    return (temp *9/5)+32

#function = c_to_f ✅
#collection = celsius_temp ✅

fahrenheit_temp  = map(c_to_f, celsius_temp)
for temp in fahrenheit_temp:
    print(temp)
```

*Formula for map function = map(function, collection)*

```python
# using lambda function in map function

celsius_temp = [0.0, 10.0, 20.0, 25.0, 30.0, 40.0, 50.0]

fahrenheit_temp  = list(map(lambda temp:(temp *9/5)+32, celsius_temp))

print(fahrenheit_temp)
```

```python
#Example 2 on map function
num = [0.0, 10.0, 20.0, 25.0, 30.0, 40.0, 50.0]

def square(x):
    return x*x

number = (map(square, num))
for i in number:
          print(i)
```

```python
num = [0.0, 10.0, 20.0, 25.0, 30.0, 40.0, 50.0]

number = list(map(lambda x:x*x, num))
print(number)
```

#### Filter Function: This function takes in function and list. 
- The filter function returns all the elements that pass a condition
- Formula for filter function = filter(function, list)

```python
# Example 1 with def function

grade = [91, 32, 83, 44, 75, 56, 67]

def is_passing(grade):
    return grade>=50

passing_grades = list(filter(is_passing, grade))
print(passing_grades, type(passing_grades))
```

```python
#Example 1 with Lambda function
grade = [91, 32, 83, 44, 75, 56, 67]

passing_grades = list(filter(lambda grade:grade>=50, grade))
print(passing_grades)
```

#### Zip Function

- zip function aggregates elements from two or more iterables such as list, tuples, sets, dictionary etc. 
- It will create a zip object with paired elements stored in tuples for each element

```python
#Example 1 using for loop with zip function

column = ['Column1', 'Column2', 'Column3']
username = ['Prof', 'Bro', 'Mister']
passwords = ('password', 'abc123', 'guest')
login_date = ['1/1/2021', '1/2/2021', '1/3/2021']
Phone_number = [1234, 5678, 9012]
Dec = [0.0, 3.2,4.4]

users  = zip(column,username, passwords, login_date, Phone_number, Dec)
for i in users:
    print(i)
```

```python
#Example 1 using list with zip function

column = ['Column1', 'Column2', 'Column3']
username = ['Prof', 'Bro', 'Mister']
passwords = ('password', 'abc123', 'guest')
login_date = ['1/1/2021', '1/2/2021', '1/3/2021']
Phone_number = [1234, 5678, 9012]
Dec = [0.0, 3.2,4.4]



users  = list(zip(column,username, passwords, login_date, Phone_number, Dec))
print(users)
```

### Reduce Function

- Reduce function: This function reduces the elements in a collection to a single value
- Formula = reduce(function, collection)
#### Note: To use reduce function, import reduce from functools

```python
from functools import reduce
```

```python
prices = [19.99, 1.00, 5.75, 12.99, 10.99, 30, 40, 50, 60, 100]

def multiply(x,y):
    return x*y

total = reduce(multiply, prices) 
print(f'${total}')
```

```python
prices = [19.99, 1.00, 5.75, 12.99, 10.99, 30, 40, 50, 60, 100]

total = reduce(lambda x,y:x*y, prices) 
print(f'${total}')
```

```python
fruit = ['Banana', 'Pawpaw', 'Pineapple', 'Apple']

def add(x,y):
    return x+y

total = reduce(add, fruit) 
print(f'${total}')
```

### List Comprehension

- This is a concise way to create lists in Python
- They are compact and easier to read than traditional loop

#### Formula = [Expression for value in iterable if condition]

```python
#Example for loop 

doubles = []
for x in range (1,11):
    doubles.append(x*2)
print(doubles)
    
```

#### List comprehension

- Formula = [Expression for value in iterable if condition]
- Formula breakdown = [Expression | for value in iterable | if condition]

```python
doubles = [x*2 for x in range(1,11)]
print(doubles)
```

```python
#Example for loop 

triples = []
for x in range(1, 6):
    triples.append(x*3)
print(triples)
```

```python
triples = [x*3 for x in range(1, 6)]
print(triples)
```

#### List comprehension

- Formula = [Expression for value in iterable if condition]
- Formula breakdown = [Expression | for value in iterable | if condition]

```python
#Example 3
pos_number = [1,-2, 3, -4, 5,-6]

for num in pos_number:
    if num >= 0:
        print(num)
```

```python
#Example for if condition in list comprehension
pos_number = [1,-2, 3, -4, 5,-6]

pos_number = [num for num in pos_number if num >= 0]
print(pos_number)
```

### Class and Objects

Class methods allow operations related to the class itself
- Class - Take (self) as the first parameter, which represents the class itself

```python
class Person:
  def __init__(self, name, age, marital_status, height, weight, colour, tribe, size):
    self.name = name
    self.age = age
    self.marital_status = marital_status
    self.height = height
    self.weight = weight
    self.colour = colour
    self.tribe = tribe
    self.size = size
      
  def __str__(self):
      return f"{self.name}, {self.age}, {self.marital_status}, {self.height}, {self.weight}, {self.colour}, {self.tribe},{self.size}"
 
p1 = Person("John", 36, 'Single', 1.76, 57, 'Brown', 'IIHY', 32)
print(p1)
p2 = Person('David', 46, 'Married', 1.76, 60, 'Fair', 'Christian', 40)
print(p2)
```

```python
class School:
    def __init__(self, course, student, classes):
        self.course = course
        self.student = student
        self.classes = classes

    def __str__(self):
      return f"{self.course}, {self.student}, {self.classes}"

S1 = School("Biology101", 'James', "100Level")
print(S1)
```

#### Try and Exception

- Expection is an event that interrupts the flow of a program
- Examples of the error include ZeroDivisionError, TypeError, ValueError etc.
- Formula = Try, Except, Finally

```python
try:
    number = int(input('Enter a number: ') )
    print(1/number)
except ZeroDivisionError:
    print("The reason you are not getting correct output is because 1 is indivisible by 0")
except ValueError:
    print('Put in numbers. Thank you!')

fruits = ['Banana', 'Orange', 'Apple']
for fruit in fruits:
    print(fruit)
```

```python
try:
    1 + "a"
except Exception:
    print('Something went wrong')
finally:
    print("Good bye!, See you tomorrow")
```

```python
#Zero Division Error

try:
    x = 10 / 0
except ZeroDivisionError as e:
    print(f"Caught an error: {e}")
```

```python
#Import Error

try:
    import non_existing_module
except ImportError as e:
    print(f"Caught an error: {e}")
```

```python
try:
    import English
except Exception:
    print('There is no module called English in Python')
```

```python
#Value Error

try:
    num = int("abc")
except ValueError as e:
    print(f"Caught an error: {e}")
```

```python
#IndexError

try:
    my_list = [1, 2, 3,4,5,6]
    print(my_list[6])
except IndexError as e:
    print(f"Caught an error: {e}")
```

```python
# File Not Found Error

try:
    with open("nonexistent.txt", "r") as file:
        data = file.read()
except FileNotFoundError as e:
    print(f"Caught an error: {e}")
```

```python
# Type Error

try:
    result = '2' + 5
except TypeError as e:
    print(f"Caught an error: {e}")
```
