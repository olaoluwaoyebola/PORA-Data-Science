```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {"A":[1,2,np.nan,4,5,7, np.nan], "B": [2,5, np.nan, np.nan,6,8, np.nan]}
```

```python
data = pd.DataFrame(data)
```

```python
print(data)
```

```python
#identifying missing values in the dataset "data" and counting it
data.isnull().sum()
```

```python
data.info()
```

```python
sns.heatmap(data.isnull(), yticklabels=False, cbar=False, cmap='viridis')
plt.show()
```

```python
data
```

```python
data["A"].fillna(data["A"].mean())
```

```python
data["B"].fillna(data["B"].median())
```

```python
df = {"A":["Sola","Sola","Ester","Dabong",np.nan,np.nan], "B": ["Male","Male","Female","Male",np.nan,np.nan]}
print(pd.DataFrame(df))
```

```python
df["B"].fillna(df["B"].mode()[0])
```
